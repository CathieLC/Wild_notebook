# Execute the code below

```
import pandas as pd
import numpy as np
link = "https://raw.githubusercontent.com/murpi/wilddata/master/quests/weather2019.csv"
df_weather = pd.read_csv(link)
```

# You have now a DataFrame with lots of weather data

```
df_weather.head()
```

# PIVOT TABLE

## It's like in Excel, but with Pandas

There are 2 syntaxes, choose your preferred :
- As a method : YOUR_DATAFRAME.pivot_table()
- As a function : pd.crosstab()

```
# Example
df_weather.pivot_table(values = 'MAX_TEMPERATURE_C', index = 'MONTH', aggfunc = 'mean')
```

## Create a Pivot table with :
- Months on rows
- the maximum of Max Temperature on values
- Opinion on columns

```
# Hop !

df_weather.pivot_table(values = 'MAX_TEMPERATURE_C', index = 'MONTH', columns='OPINION', aggfunc = np.amax )
```

## Pivot table with :
- Days on rows
- Precipitation on values
- Months on columns


```
# Hop !
df_weather.pivot_table(values = 'PRECIP_TOTAL_DAY_MM', columns='MONTH', index = 'DAY')

```

## Pivot table with :
- Months on rows
- the sum of Precipitations on values

**Then display it as a column chart.**

```
# Hop !
df_weather.pivot_table(values = 'PRECIP_TOTAL_DAY_MM', index = 'MONTH', aggfunc = np.sum)
```

## Pivot table with :
- Months on rows
- the average of Max Temperature on a first column of values
- the average of Sunhours on a second column of values

```
# Hop !

pivot_table_graph = df_weather.pivot_table(values = ['MAX_TEMPERATURE_C', 'SUNHOUR'],
                       index = 'MONTH',
                       aggfunc = {'MAX_TEMPERATURE_C' : np.mean,
                                  'SUNHOUR' : np.mean})
pivot_table_graph
```

# Chart
Using your favorite library (MatPlotLib, Seaborn, Plotly or what you want), draw a chart from the previous pivot table with :
- Temperature in a blue column chart
- Sunhours in a red line plot

Tip : Once your pivot table created, you can help yourself with the `.reset_index()`  method.

```
pivot_table_graph.reset_index(inplace=True)
pivot_table_graph
```

```
# Your code here
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd


ax = pivot_table_graph.plot(x='MONTH', y="MAX_TEMPERATURE_C", kind='bar', figsize=(10,6), label='MAX_TEMP', ylabel="Tempratures")
pivot_table_graph.plot(x='MONTH', y="SUNHOUR", kind='line', ax=ax, use_index=False, c='red',)

plt.show()





```
