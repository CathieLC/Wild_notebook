```python
import pandas as pd

link = 'https://raw.githubusercontent.com/justmarkham/pandas-videos/master/data/imdb_1000.csv'
df = pd.read_csv(link)
```

```python
# Map used for substituting each value in a Series with another value, here replace series by str

df['actors_list'] = df['actors_list'].map(str)
```

```python
#create a loop for cleaning base with replace [ and u and ' and ]...

for i in range(len(df['actors_list'])):
    df['actors_list'][i] = df['actors_list'][i].replace("[u\'", "").replace("u\'", "").replace("\'", '').replace("]","").replace("u\"", "").replace("\"", "")
    
df
```

```python
# split before explode

df['actors_list'] = df['actors_list'].str.split(",")
df
```

```python
# explode for one row with one value of actors_list column

df = df.explode('actors_list')
df
```

```python
#reset index for df

df = df.reset_index(drop=True)
df
```
