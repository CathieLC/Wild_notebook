# La méthode `Cross Validation`

* Le jeu de données total est divisé plusieurs parties.
* Pour chaque partie, un sous-modèle est entraîné et évalué.
* Le score moyen de tous les sous-modèles devient le score à prendre en compte pour évaluer le modèle.

Il faut déterminer le nombre de splits, via l'`hyperparamètre cv`. Par défaut, `cv = 5`      
Une bonne pratique est de garder `cv = 5` lorsque qu'on a un petit jeu de données (- de 2000 observations).   
On peut se permettre `cv = 10` si on a un grand jeu de données (+ de 2000 observations).

⚠️ La `cross validation` ne produit pas un modèle entraîné,  
elle ne note qu'un modèle hypothétique entraîné sur l'ensemble des données.
