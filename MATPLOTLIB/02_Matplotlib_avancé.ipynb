```python
import matplotlib.pyplot as plt
import pandas as pd

fig, axs = plt.subplots(2, 2, figsize=(23, 16))

#Scatterplot
axs[0,0].scatter(x = df_weather["DATE"], y = df_weather["MAX_TEMPERATURE_C"], c= 'indigo', label = "Max")
axs[0,0].scatter(x = df_weather["DATE"], y = df_weather["MIN_TEMPERATURE_C"], c= 'brown', label = "Min")
axs[0,0].set_title("Weather Max 2019", size = 16, c="purple",  loc = "right")
axs[0,0].set_xlabel("Janvier 2019 à Janvier 2020", size = 13)
axs[0,0].set_ylabel("Temperatures", size = 13)
axs[0,0].legend(title = "Temperatures", loc = "upper left")



#PieChart
opinions = 'very bad','bad', 'good', 'not good not bad', 'very good'
percent = list(df_weather['OPINION'].value_counts())
axs[0,1].pie(percent, labels=opinions, autopct='%1.1f%%')
axs[0,1].set_title("Value count - % Opinions", size = 16, c="purple",  loc = "right")




# column chart and curve
month_and_temp_max_of_January = pd.read_csv(link).iloc[:,[0,1,2]].head(31)
month_and_temp_max_of_January['DATE'] = pd.to_datetime(month_and_temp_max_of_January['DATE'])

axs[1,0].bar(month_and_temp_max_of_January['DATE'], month_and_temp_max_of_January['MAX_TEMPERATURE_C'], label = "MAX")
axs[1,0].plot(month_and_temp_max_of_January['DATE'], month_and_temp_max_of_January['MIN_TEMPERATURE_C'], c="red", label = "MIN")
axs[1,0].set_title("Janvier 2019 : Températures max VS températures min", size = 16, c="purple",  loc = "right")
axs[1,0].set_xlabel("Dates/Janvier2019", size = 13)
axs[1,0].legend(title = "Temperatures", loc = "upper left")




# Histogramme
axs[1,1].hist(df_weather["MAX_TEMPERATURE_C"])
axs[1,1].set_title("Nb de jours du max des températures relevées", size = 16, c="purple",  loc = "right")
axs[1,1].set_xlabel("Températures", size = 13)
axs[1,1].set_ylabel("Nombres de jours", size = 13)
```

Could you draw 1 scatterplot with :

- Colorize in green the good and very good points
- Colorize in red the bad and very bad points
- Colorize in yellow the not good not bad points

Tips :

```python
# Je reprends mon tableau avec mes valeurs et mes 4 colonnes

df_weather = pd.read_csv("https://raw.githubusercontent.com/murpi/wilddata/master/quests/weather2019.csv").iloc[:,[0, 1, 2, -3]]
df_weather
```

```python
# creation de 3 df avec good very good etc.

df_weather['DATE'] = pd.to_datetime(df_weather['DATE'])

gf_gvg = df_weather.loc[(df_weather["OPINION"]=="good") | (df_weather["OPINION"]=="very good")]
df_nbng = df_weather.loc[df_weather["OPINION"]=="not good not bad"]
df_bvb = df_weather.loc[(df_weather["OPINION"]=="bad") | (df_weather["OPINION"]=="very bad")]
```

```python
#création des 3 graphs

fig = plt.figure(figsize = (10, 5))


plt.scatter(x = gf_gvg["DATE"] , y = gf_gvg["MAX_TEMPERATURE_C"], c= 'green', label = "Good Very Good")
plt.scatter(x = df_nbng["DATE"] , y = df_nbng["MAX_TEMPERATURE_C"], c= 'yellow', label = "Not bad not good")
plt.scatter(x = df_bvb["DATE"] , y = df_bvb["MAX_TEMPERATURE_C"], c= 'red', label = "Bad Very Bad")

plt.title("Temperature Evolution by date", size = 16, c="black",  loc = "center")
plt.xlabel("DATE", size = 13)
plt.ylabel("MAX Temperature", size = 13)
plt.legend( loc = "upper right")

plt.show()
```
