# Mission 3

Crée une fonction qui prend en paramètre un dataframe, et affiche le taux de remplissage de chacune de ces colonnes sous forme de pourcentage. C'est à dire :
- colonne 1 : 73%
- colonne 2 : 99%

```
# calculer les pourcentages d'un dataframe par colonne
tx_remplissage = pd.DataFrame(round(df.notna().mean()*100,2), columns=['percent'])
```

```
tx_remplissage
```

# Mission 4

Quel est le nombre de colonne appartenant à chaque quartile des taux de remplissage ? Autrement dit, combien de colonne sont remplies à 0%, entre 0 et 25%, entre 25 et 50%...

```
dfzero = tx_remplissage[tx_remplissage['percent']==0].count().sum()
df_0_25 = tx_remplissage[(tx_remplissage['percent'] >0) & (tx_remplissage['percent'] < 25) ].count().sum()
df_25_50 = tx_remplissage[(tx_remplissage['percent'] >= 25) & (tx_remplissage['percent'] < 50) ].count().sum()
df_50_75 = tx_remplissage[(tx_remplissage['percent'] >= 50) & (tx_remplissage['percent'] < 75) ].count().sum()
df_75_100 = tx_remplissage[(tx_remplissage['percent'] >= 75) & (tx_remplissage['percent'] < 100) ].count().sum()

print(f"le nombre de colonne appartenant à chaque quartile est \n 0 % = {dfzero} \n Entre 0 % et 25 % = {df_0_25}\n Entre 25 % et 50 % = {df_25_50}\n Entre 50 % et 75 % = {df_50_75} \n Entre 75 % et 100 % = {df_75_100} ")
```

# Mission 6

Un dataset trop pauvre en data n'est pas vraiment intéressant. Après avoir analyser les colonnes, nous nous sommes rendu compte que nous pouvons supprimer les colonnes avec moins de 60% de valeurs non nulles, sauf les colonnes pnns_groups_1 et pnns_groups_2, nutrition-score-fr_100g et nutrition-score-uk_100g. En effet ces 2 colonnes vont nous être utiles par la suite.

* creator

```
tx_remp_country_fr = pd.DataFrame(round(countries_france.notna().mean()*100,2), columns=['percent'])
tx_remp_country_fr.reset_index(inplace= True)
tx_remp_country_fr.rename(columns={'index': "name_column"}, inplace=True)

#exclure ces colonnes : pnns_groups_1 et pnns_groups_2, nutrition-score-fr_100g et nutrition-score-uk_100g => elles sont toutes à plus de 60 %

df_filtered = tx_remp_country_fr[tx_remp_country_fr['percent'] >= 60]
df_filtered
```
